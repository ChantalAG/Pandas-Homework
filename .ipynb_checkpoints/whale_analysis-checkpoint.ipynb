## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for for all portfolios using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
portfolio_returns_df.rolling(window=21).std()
```

```python
# Plot the rolling standard deviation
portfolio_returns_df.rolling(window=21).std().plot(figsize=(15,10), title="All Portfolios Rolling Standard Deviation")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = portfolio_returns_df.corr()


# Display de correlation matrix
correlation
```

### Determine which portfolio may mimick the S&P TSX 60.
Algo 2 has a correlation of 0.73730 with S&P TSX 60, and has the highest positive correlation compared to the other portfolios, therefore it may mimick the S&P TSX 60. 

```python
import seaborn as sns
sns.heatmap(correlation, vmin=-1, vmax=1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio (Algo 1)
cov_algo_1 = portfolio_returns_df['Algo 1'].cov(portfolio_returns_df['S&P TSX 60'])

cov_algo_1
```

```python
# Calculate variance of S&P TSX
var_sp_tsx = portfolio_returns_df['S&P TSX 60'].var()
var_sp_tsx
```

```python
# Computing beta
beta = cov_algo_1/var_sp_tsx

beta
```

### Determine Volatility
Since beta < 1 but > 0 Algo 1 is less volatile than the S&P TSX 60 market. The rolling 60 day beta is between -0.2 and 0.6, indicating that Algo 1 fluctuates between being negatively correlated with the market and less volatile than the market.

```python
# Plot beta trend (60 day rolling beta for Algo 1 returns and S&P TSX 60)
rolling_cov = portfolio_returns_df['Algo 1'].rolling(window=60).cov(portfolio_returns_df['S&P TSX 60'])
rolling_var = portfolio_returns_df['S&P TSX 60'].rolling(window=60).var()
rolling_beta = rolling_cov/rolling_var
rolling_beta
```

```python
rolling_beta.plot(figsize = (15,10), title="Rolling 60 day Beta: Algo 1 vs S&P TSX 60")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
ewm = portfolio_returns_df.ewm(halflife=21).std()
ewm.head()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_r=portfolio_returns_df.mean()*252/portfolio_returns_std*np.sqrt(252)
sharpe_r
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_r.plot.bar(figsize = (10,5), title="Annualized Sharpe Ratios")
```

### Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.

The higher the sharpe ratio the better the risk-adjusted performance. Both Algo 1 and Algo 2 outperforms the market (S&P TSX 60) and the whales portfolios, with one exception: The BERKSHIRE HATHAWAY INC has a higher sharpe ratio when compared to Algo 2. However, overall you can conclude that the algorithmic strageties outperfom the market and the whales portfolios.
